```python
# Import libraries and custom modules
import pandas as pd
from heart_disease_predictor import load_data, preprocess_data, train_model, evaluate_model, display_feature_importance
from eda_visuals import plot_class_distribution, plot_correlation_heatmap
```

```python
# Load data
df = load_data(r"C:\Users\AAFIA\Desktop\heart_disease_prediction\Heart_disease_cleveland_new.csv")
df.head()
```

```python
# Clean the dataset: handle missing values
def clean_dataset(df):
    print("üîç Missing values in each column:")
    print(df.isnull().sum())
    print("\n‚úÖ Total missing values:", df.isnull().sum().sum())
    df_clean = df.dropna()
    print(f"\nüßπ Cleaned dataset shape: {df_clean.shape}")
    return df_clean

df = clean_dataset(df)
```

```python
# EDA: Plot class distribution
plot_class_distribution(df)
```

```python
# EDA: Correlation heatmap
plot_correlation_heatmap(df)
```

```python
# Preprocess data for training
X_train, X_test, y_train, y_test, scaler = preprocess_data(df)
```

```python
# Train classification model (logistic or decision tree)
model = train_model(X_train, y_train, model_type='logistic')  # Change to 'tree' if needed
```

```python
# Evaluate the model
evaluate_model(model, X_test, y_test)
```
